```python
def f_LDA(data_x, data_y):
    f_x = StandardScaler().fit_transform(data_x)
    f_lda = LinearDiscriminantAnalysis()
    f_result = f_lda.fit(f_x, data_y)
    f_pred = f_lda.predict(f_x)
    f_pred_prob = f_lda.predict_proba(f_x)[:,1]
    f_intercept = lda_result.intercept_
    
    z = 0
    d = 0
    z_value = []
    d_score = []

    for i in f_x:
        for j in range(5):
            z += (lda_result.coef_[0,j])*(i[j])
        else:
            d = z + f_intercept
            z_value.append(z)
            d_score.append(float(d))
    
    f_data_result = pds.DataFrame(data_x, columns=['satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company'])
    #f_data_result['left'] = y
    f_data_result['pred_prob'] = f_pred_prob
    f_data_result['pred'] = f_pred
    f_data_result['d_score'] = d_score
    f_data_result['d_class'] = ""
    f_data_result['z_value'] = z_value
    
    return f_data_result
```

```python
x = data_hr.drop(columns='left')
y = data_hr['left']

_result = f_LDA(x, y)

d_class = []
z_class = []

for i in _result['z_value']:
    if (i < 0.5): z_class.append(0)
    elif (i >= 0.5): z_class.append(1)

for i in _result['d_score']:
    if (i < 0.5): d_class.append(0)
    elif (i >= 0.5): d_class.append(1)

_result['d_class'] = d_class
_result['z_class'] = z_class

_result.round(8)
```
