```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

data = pd.read_csv('../input/19560-indian-takeaway-orders/restaurant-1-orders.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().count()
```

```python
data_basket = (data.groupby(['Order Number', 'Item Name'])['Quantity']
              .sum().unstack().reset_index().fillna(0)
              .set_index('Order Number'))
```

```python
data_basket.head()
```

```python
data_basket.shape
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
data_basket.columns
```

```python
data_basket_sets = data_basket.applymap(encode_units)

data_basket_sets
```

```python
data_basket_sets.shape
```

```python
freq_itemset = apriori(data_basket_sets, min_support=0.05, use_colnames=True)
freq_itemset.shape
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset, metric="lift", min_threshold=1)
rules.head()
```

```python
sort_confidence = rules.sort_values(by=['confidence'], ascending=False)
sort_confidence.head(10)
```

```python
sort_lift = rules.sort_values(by=['lift'], ascending=False)
sort_lift.head(10)
```

```python
sort_conf_lift = rules.sort_values(by=['confidence', 'lift'], ascending=False)
sort_conf_lift.head()
```
