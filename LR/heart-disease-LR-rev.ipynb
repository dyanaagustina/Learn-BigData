```python
def f_logit(x_sig, y_sig):
    f_x = StandardScaler().fit_transform(x_sig)
    f_reg = LogisticRegression()
    f_log_reg = _reg.fit(f_x, y_sig)
    f_pred = _reg.predict(f_x)
    f_pred_prob = _reg.predict_proba(f_x)[:,1]
    
    f_intercept = f_log_reg.intercept_
    f_coef = f_log_reg.coef_
    
    y = 0
    prob = 0
    y_prob = []
    
    for i in f_x:
        for j in range(0,7):
            y += f_coef[0,j] * i[j]
        else:
            prob = 1/(1+npy.exp(-(f_intercept + y)))
            y_prob.append(float(prob))
    
    data_result = pds.DataFrame(x_sig, columns=['sex','cp','thalac','exang','oldpeak','ca','thal'])
    data_result['pred_prob'] = f_pred_prob
    data_result['pred'] = f_pred
    data_result['y_prob'] = y_prob
     
    return data_result
```

```python
x = data_heart.drop(columns=['age', 'restbps', 'chol', 'fbs', 'restecg', 'slope','target'])
y = data_heart['target']

_result = f_logit(x, y)

y_pred = []

for i in _result['y_prob']:
    if (i < 0.5): y_pred.append(0)
    elif (i >= 0.5): y_pred.append(1)

_result['y_pred'] = y_pred
_result
```
