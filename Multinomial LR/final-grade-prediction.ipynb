```python
def famrel_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,0]
    elif cat == 2: k = log_reg_result.coef_[1,0]
    elif cat == 3: k = log_reg_result.coef_[2,0]
    elif cat == 4: k = log_reg_result.coef_[3,0]
    else: k = log_reg_result.coef_[4,0]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
famrel_koef(3)
```

```python
def freetime_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,1]
    elif cat == 2: k = log_reg_result.coef_[1,1]
    elif cat == 3: k = log_reg_result.coef_[2,1]
    elif cat == 4: k = log_reg_result.coef_[3,1]
    else: k = log_reg_result.coef_[4,1]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
freetime_koef(3)
```

```python
def goout_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,2]
    elif cat == 2: k = log_reg_result.coef_[1,2]
    elif cat == 3: k = log_reg_result.coef_[2,2]
    elif cat == 4: k = log_reg_result.coef_[3,2]
    else: k = log_reg_result.coef_[4,2]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
goout_koef(3)
```

```python
def dalc_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,3]
    elif cat == 2: k = log_reg_result.coef_[1,3]
    elif cat == 3: k = log_reg_result.coef_[2,3]
    elif cat == 4: k = log_reg_result.coef_[3,3]
    else: k = log_reg_result.coef_[4,3]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
dalc_koef(3)
```

```python
def walc_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,4]
    elif cat == 2: k = log_reg_result.coef_[1,4]
    elif cat == 3: k = log_reg_result.coef_[2,4]
    elif cat == 4: k = log_reg_result.coef_[3,4]
    else: k = log_reg_result.coef_[4,4]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
walc_koef(3)
```

```python
def health_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,5]
    elif cat == 2: k = log_reg_result.coef_[1,5]
    elif cat == 3: k = log_reg_result.coef_[2,5]
    elif cat == 4: k = log_reg_result.coef_[3,5]
    else: k = log_reg_result.coef_[4,5]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
health_koef(3)
```

```python
def g1_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,6]
    elif cat == 2: k = log_reg_result.coef_[1,6]
    elif cat == 3: k = log_reg_result.coef_[2,6]
    elif cat == 4: k = log_reg_result.coef_[3,6]
    else: k = log_reg_result.coef_[4,6]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
g1_koef(3)
```

```python
def g2_koef(cat):
    
    if cat == 1: k = log_reg_result.coef_[0,7]
    elif cat == 2: k = log_reg_result.coef_[1,7]
    elif cat == 3: k = log_reg_result.coef_[2,7]
    elif cat == 4: k = log_reg_result.coef_[3,7]
    else: k = log_reg_result.coef_[4,7]
   
    #print(k)
    #print(k*cat)
    return k*cat
        
g2_koef(3)
```

```python
def persamaan(cat,famrel_cat,freetime_cat,goout_cat,dalc_cat,walc_cat,health_cat,g1_cat,g2_cat):
    if cat == 1: i = log_reg_result.intercept_[0]
    elif cat == 2: i = log_reg_result.intercept_[1]
    elif cat == 3: i = log_reg_result.intercept_[2]
    elif cat == 4: i = log_reg_result.intercept_[3]
    elif cat == 5: i = log_reg_result.intercept_[4]
    
    famrel = famrel_koef(famrel_cat)
    freetime = freetime_koef(freetime_cat)
    goout = goout_koef(goout_cat)
    dalc = dalc_koef(dalc_cat)
    walc = walc_koef(walc_cat)
    health = health_koef(health_cat)
    g1 = g1_koef(g1_cat)
    g2 = g2_koef(g2_cat)
    
    pers = i + famrel + freetime + goout + dalc + walc + health + g1 + g2
    
    return pers
    
persamaan(3,1,4,4,2,3,1,2,1)
```

```python
def nilai_y(famrel_cat,freetime_cat,goout_cat,dalc_cat,walc_cat,health_cat,g1_cat,g2_cat):
    y_prob_1 = 1/(1+npy.exp(persamaan(1,famrel_cat,freetime_cat,goout_cat,dalc_cat,walc_cat,health_cat,g1_cat,g2_cat)))
    y_prob_2 = 1/(1+npy.exp(persamaan(2,famrel_cat,freetime_cat,goout_cat,dalc_cat,walc_cat,health_cat,g1_cat,g2_cat)))
    y_prob_3 = 1/(1+npy.exp(persamaan(3,famrel_cat,freetime_cat,goout_cat,dalc_cat,walc_cat,health_cat,g1_cat,g2_cat)))
    y_prob_4 = 1/(1+npy.exp(persamaan(4,famrel_cat,freetime_cat,goout_cat,dalc_cat,walc_cat,health_cat,g1_cat,g2_cat)))
    y_prob_5 = 1/(1+npy.exp(persamaan(5,famrel_cat,freetime_cat,goout_cat,dalc_cat,walc_cat,health_cat,g1_cat,g2_cat)))
    
    y_prob = [y_prob_1,y_prob_2,y_prob_3,y_prob_4]

    return y_prob

nilai_y(3,3,3,3,2,2,2,2)
```
