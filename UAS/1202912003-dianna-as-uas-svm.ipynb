Melakukan import library yang akan digunakan dalam pemodelan data menggunakan SVM. Kemudian load dataset ke dalam DataFrame pandas.

```python
import pandas as pd
import matplotlib.pyplot as plb
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('../input/uassvmdata/UAS_SVM_Dianna.csv')
data.head()
```

Melakukan pengecekan dataset untuk mengetahui apakah terdapat kolom yang bernilai data null.

```python
data.isnull().sum()
```

Menentukan variabel dependen dan independen. Untuk variabel dependen adalah kolom "Pelunasan Kredit", sementara kolom lainnya merupakan variabel independen.

Data kemudian dibagi menjadi data training dan data testing dengan komposisi data training sebesar 70% dan data testing 30%.

Lakukan normalisasi data menggunakan fungsi StandardScaler() pada variabel independen.

```python
x = data.drop(columns=['Pelunasan Kredit'])
y = data['Pelunasan Kredit']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)

x_test_std
```

Buat pemodelan menggunakan fungsi SVC pada library SVM. Kemudian pemodelan tersebut diaplikasikan terhadap data training dengan variabel independen yang sudah dinormalisasi.

```python
svc = svm.SVC(C=1.0, random_state=1, kernel='linear')
 
# Fit the model
svm_result = svc.fit(x_train_std, y_train)
```

Lakukan pengetesan untuk menghasilkan prediksi dengan menggunakan fungsi predict() terhadap data testing.

```python
y_predict = svc.predict(x_test_std)
y_predict
```

```python
Kombinasikan hasil data prediksi dari pemodelan data dengan data testing sebenarnya menjadi sebuah DataFrame.
```

```python
data_result = pd.DataFrame(x_test, columns=x_test.columns)
data_result['Hasil SVM'] = y_predict
data_result
```

Hitung nilai akurasi pemodelan terhadap data testing 30%. Pemodelan menggunakan SVM ini menghasilkan akurasi sebesari 0.497 ~ 49,7%

```python
print("Accuracy score %.3f" %metrics.accuracy_score(y_test, y_predict))
```
